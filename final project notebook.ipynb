```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
top_customers = df.groupby('customer_id')['payment_value'].sum().nlargest(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_customers.index, y=top_customers.values, palette="Blues_r")
plt.xticks(rotation=90)
plt.xlabel("Customer ID")
plt.ylabel("Total Spending")
plt.title("Top 10 Spending Customers")
plt.show()
```

```python
state_orders = df['customer_state'].value_counts()

plt.figure(figsize=(12, 5))
sns.barplot(x=state_orders.index, y=state_orders.values, palette="coolwarm")
plt.xlabel("State")
plt.ylabel("Number of Orders")
plt.title("Orders by State")
plt.xticks(rotation=0)
plt.show()
```

```python
top_products = df['product_id'].value_counts().nlargest(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_products.index, y=top_products.values, palette="viridis")
plt.xticks(rotation=90)
plt.xlabel("Product ID")
plt.ylabel("Number of Orders")
plt.title("Top 10 Best-Selling Products")
plt.show()
```

```python
top_products = df['product_category_name'].value_counts().nlargest(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_products.index, y=top_products.values, palette="viridis")
plt.xticks(rotation=90)
plt.xlabel("Product ID")
plt.ylabel("Number of Orders")
plt.title("Top 10 Best-Selling Products")
plt.show()
```

```python
category_reviews = df.groupby('product_category_name')['review_score'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 5))
sns.barplot(x=category_reviews.index, y=category_reviews.values, palette="rocket")
plt.xticks(rotation=90)
plt.xlabel("Product Category")
plt.ylabel("Average Review Score")
plt.title("Top 10 Product Categories by Review Score")
plt.show()
```

```python
top_sellers = df['seller_id'].value_counts().nlargest(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_sellers.index, y=top_sellers.values, palette="magma")
plt.xticks(rotation=90)
plt.xlabel("Seller ID")
plt.ylabel("Number of Orders")
plt.title("Top 10 Sellers by Sales")
plt.show()
```

```python
df['delivery_time'] = (pd.to_datetime(df['order_delivered_customer_date']) - pd.to_datetime(df['order_purchase_timestamp'])).dt.days
shipping_efficiency = df.groupby('seller_id')['delivery_time'].mean().sort_values().head(10)

plt.figure(figsize=(12, 5))
sns.barplot(x=shipping_efficiency.index, y=shipping_efficiency.values, palette="coolwarm")
plt.xticks(rotation=90)
plt.xlabel("Seller ID")
plt.ylabel("Average Delivery Time (Days)")
plt.title("Top 10 Sellers with Fastest Delivery")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['delivery_time'].dropna(), bins=30, kde=True, color="purple")
plt.xlabel("Delivery Time (Days)")
plt.ylabel("Frequency")
plt.title("Distribution of Delivery Times")
plt.show()
```

```python
df['late_delivery'] = df['order_delivered_customer_date'] > df['order_estimated_delivery_date']
late_deliveries = df['late_delivery'].value_counts()

plt.figure(figsize=(6, 5))
sns.barplot(x=late_deliveries.index, y=late_deliveries.values, palette="Reds")
plt.xticks(ticks=[0, 1], labels=["On Time", "Late"])
plt.xlabel("Delivery Status")
plt.ylabel("Number of Orders")
plt.title("Late vs On-Time Deliveries")
plt.show()
```

```python
df['hour'] = pd.to_datetime(df['order_purchase_timestamp']).dt.hour 

# Group by hour
hourly_sales = df.groupby('hour').size().reset_index(name='count')

# Plot
plt.figure(figsize=(10, 5))
plt.bar(hourly_sales['hour'], hourly_sales['count'], color='black')
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Orders")
plt.title("Orders per Hour")
plt.xticks(range(24))  # Show all hours from 0 to 23
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
payment_types = df['payment_type'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=payment_types.index, y=payment_types.values, palette="Set2")
plt.xlabel("Payment Method")
plt.ylabel("Number of Transactions")
plt.title("Popular Payment Methods")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['payment_installments'], bins=20, kde=True, color="teal")
plt.xlabel("Number of Installments")
plt.ylabel("Frequency")
plt.title("Distribution of Payment Installments")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x=df['review_score'], palette="Spectral")
plt.xlabel("Review Score")
plt.ylabel("Number of Reviews")
plt.title("Distribution of Review Scores")
plt.show()
```

```python
df['review_length'] = df['review_comment_message'].astype(str).apply(len)

plt.figure(figsize=(10, 5))
sns.boxplot(x=df['review_score'], y=df['review_length'], palette="Blues")
plt.xlabel("Review Score")
plt.ylabel("Review Length (Characters)")
plt.title("Review Length vs Review Score")
plt.show()
```

```python
df = df.apply(lambda x: x.fillna(x.mean()), axis=0)
```

```python

# Assuming df is your DataFrame
for df['review_comment_title'] in df.columns:
    # Check if the column has any NaN values
    if df['review_comment_title'].isnull().sum() > 0:
        # Calculate the mean of the column (ignoring NaN values)
        mean_value = df['review_comment_title'].mean()
        # Replace NaN with the calculated mean
        df['review_comment_title'].fillna(mean_value, inplace=True)

# Check the result
print(df)
```
